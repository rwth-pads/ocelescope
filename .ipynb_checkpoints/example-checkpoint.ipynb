# TOTeM Model Miner Example

```python
from datetime import datetime
from ocpa.objects.log.importer.ocel2.sqlite import factory as ocel_import_factory
from algorithm import mine_totem

# import the object centric event log using the ocpa library
filename = "./example-data/order-management.sqlite"
ocel = ocel_import_factory.apply(filename)
print(len(ocel.process_executions))
```

One can mine the TOTeM model for an object centric event log using the function call mine_totem.
The first parameter is the object-centric event log and the second parameter is the tau value.
With a tau value of 1 the mined model matches the log perfectly.
A lower tau value allows for noise and filters out infrequent behavior.

```python
print(len(ocel.process_executions))
```

```python
mine_totem(ocel, tau=0.9)
```
